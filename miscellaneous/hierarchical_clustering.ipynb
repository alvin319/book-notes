```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

%matplotlib inline
```

```python
dimensionality = 2
coordinate = 4
first = np.array([0, 0])
second = np.array([coordinate, coordinate])
third = np.array([0, coordinate])

num_points = 900
X = np.zeros((num_points, dimensionality))
X[:300, :] = np.random.randn(300, dimensionality) + first
X[300:600, :] = np.random.randn(300, dimensionality) + second
X[600:, :] = np.random.randn(300, dimensionality) + third
```

```python
Z = linkage(X, 'ward')
print('Z.shape {}'.format(Z.shape))
plt.title('Ward')
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'single')
print('Z.shape {}'.format(Z.shape))
plt.title('Ward')
dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'complete')
print('Z.shape {}'.format(Z.shape))
plt.title('Ward')
dendrogram(Z)
plt.show()
```
